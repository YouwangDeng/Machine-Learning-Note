# 导入GraphLab Create

```python
import graphlab
```

# 使用CIFAR-10数据集

```python
image_train = graphlab.SFrame('image_train_data/')
```

```python
deep_learning_model = graphlab.load_model('http://s3.amazonaws.com/GraphLab-Datasets/deeplearning/imagenet_model_iter45')
```

```python
deep_learning_model.extract_features(image_train)
```

# 通过应用这些深层特征来构建最近邻模型来进行图像检索

```python
knn_model = graphlab.nearest_neighbors.create(image_train,features=['deep_features'],label='id')
```

# 通过应用深度特征的图像检索模型找到相似的图像

```python
cat = image_train[18:19]
```

```python
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
cat1 = image_train[image_train['id'] == 6910]
```

```python
cat1['image'].show()
```

```python
def get_image_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')
```

```python
cat_neighbors = get_image_from_ids(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

# 找到和轿车相似的图像

```python
car = image_train[8:9]
```

```python
car_neighbors = get_image_from_ids(knn_model.query(car))
```

```python
car_neighbors['image'].show()
```

# 构造一个lambda函数来寻找和显示最近的图像

```python
show_neighbors = lambda i: get_image_from_ids(knn_model.query(image_train[i:i+1]))['image'].show()
```

```python
show_neighbors(26)
```
